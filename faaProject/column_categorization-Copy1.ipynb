Categorizes all of the columns in two kilos into one of these types:
- na: at least 95% empty or have all the same value
- text: sentence or sentence-like
- date: dates
- code: the entries are some kind of code

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./Subsets/Maintenance_Orig_data_renamed.csv', encoding="latin-1")
num_rows = len(data)
```

```python
# creates cols dictionary, which I use for the rest of the notebook as a quick reference to get info on the column
# Keys: column names
# Values: sub-dictionaries of all the unique values and the number of times those values occur

cols = {}

for col in data.columns:
    
    cols[col] = {}
    
    for value in data[col].dropna():
        cols[col][value] = cols[col].get(value, 0) + 1
```

```python
# categorizing based on values or column name

na_cols = []
date_cols = []
type_cols = []
text_cols = []
code_cols = []

for col_name in cols:
    if len(cols[col_name]) <= 1 or sum(cols[col_name].values()) < 0.05*num_rows:
        na_cols.append(col_name)
    elif "Date" in col_name or "date" in col_name or "time" in col_name or "Time" in col_name:
        date_cols.append(col_name)
    elif "Type" in col_name or "type" in col_name:
        type_cols.append(col_name)
    elif "text" in col_name or "remark" in col_name:
        text_cols.append(col_name)
    elif "Code" in col_name or "code" in col_name:
        code_cols.append(col_name)
```
